# Step 0 Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

# Step 1 Importing Data and Data Preprocessing 

```python
test_feat = pd.read_csv("test_set_features.csv")
train_label= pd.read_csv("training_set_labels.csv")
train_feat=pd.read_csv("training_set_features.csv")
```

```python

#splitting the data


train_data = pd.merge(train_feat, train_label, on='respondent_id')

# Separate features and target variables
X = train_data.drop(columns=['respondent_id', 'xyz_vaccine', 'seasonal_vaccine'])
y_xyz = train_data['xyz_vaccine']
y_seasonal = train_data['seasonal_vaccine']
X_test = test_feat.drop(columns=['respondent_id'])

```

```python
# Identifying different types of data we have 

#Categories 
categorical_cols = X.select_dtypes(include=['object']).columns

#Numerical Binary Columns - 0 or 1
binary_cols = [col for col in X.select_dtypes(include=['int64', 'float64']).columns if X[col].nunique() == 2]

#Normal Numerical Columns 
continuous_numerical_cols = [col for col in X.select_dtypes(include=['int64', 'float64']).columns if X[col].nunique() > 2]


```

# Step 2 Defining Preprocessing pipelines for ease

```python

continuous_numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

binary_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value=0))
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Combining all the pipelines

preprocessor = ColumnTransformer(
    transformers=[
        ('num', continuous_numerical_pipeline, continuous_numerical_cols),
        ('bin', binary_pipeline, binary_cols),
        ('cat', categorical_pipeline, categorical_cols)
    ])
```

# Step 3 Defining our models 


### Used LogisticReg

```python

model_xyz = LogisticRegression(max_iter=1000, random_state=42)
model_seasonal = LogisticRegression(max_iter=1000, random_state=42)


# Create separate pipelines for each target variable so that we can train them seperately

pipeline_xyz = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', model_xyz)
])

pipeline_seasonal = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', model_seasonal)
])
```

```python
# Split data into training and validation sets for xyz
X_train_xyz, X_val_xyz, y_train_xyz, y_val_xyz = train_test_split(X, y_xyz, test_size=0.2, random_state=42)

# Train and predict for xyz
pipeline_xyz.fit(X_train_xyz, y_train_xyz)
y_val_pred_prob_xyz = pipeline_xyz.predict_proba(X_val_xyz)[:, 1]
roc_auc_xyz = roc_auc_score(y_val_xyz, y_val_pred_prob_xyz,average='macro')
```

```python
# Split data into training and validation sets for seasonal
X_train_seasonal, X_val_seasonal, y_train_seasonal, y_val_seasonal = train_test_split(X, y_seasonal, test_size=0.3, random_state=42)

# Train and predict for seasonal
pipeline_seasonal.fit(X_train_seasonal, y_train_seasonal)
y_val_pred_prob_seasonal = pipeline_seasonal.predict_proba(X_val_seasonal)[:, 1]
roc_auc_seasonal = roc_auc_score(y_val_seasonal, y_val_pred_prob_seasonal,average='macro')
```

## Scores ROC AUC

```python
print(f"ROC AUC for xyz_vaccine: {roc_auc_xyz:.4f}")
print(f"ROC AUC for seasonal_vaccine: {roc_auc_seasonal:.4f}")

print(f"Mean ROC AUC : {(roc_auc_seasonal+roc_auc_xyz)/2:.4f}")
```

```python
# Retrain the models on the entire training set
pipeline_xyz.fit(X, y_xyz)
pipeline_seasonal.fit(X, y_seasonal)

# Predict on the test set
y_test_pred_xyz = pipeline_xyz.predict_proba(X_test)[:, 1]
y_test_pred_seasonal = pipeline_seasonal.predict_proba(X_test)[:, 1]
```

# Step 4 Making the submission

```python
# Prepare submission
submission = pd.DataFrame({
    'respondent_id': test_feat['respondent_id'],
    'xyz_vaccine': y_test_pred_xyz,
    'seasonal_vaccine': y_test_pred_seasonal
})

# Save to CSV
submission.to_csv('submission.csv', index=False)
```
